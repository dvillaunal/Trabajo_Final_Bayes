# De factor a Numérica (Ordinal)

```python
def replace_class(c):
    """
    Esta función remplaza los valores por numericos
    """
    if c == "A":
        return 1
    elif c == "B":
        return 2
    elif c == "C":
        return 3
    elif c == "D":
        return 4
    elif c == "E":
        return 5
    elif c == "F":
        return 6

def replace_penumbra(p):
    """
    Esta función remplaza los valores por numericos
    """
    if p == "X":
        return 1
    elif p == "R":
        return 2
    elif p == "S":
        return 3
    elif p == "A":
        return 4
    elif p == "H":
        return 5
    elif p == "K":
        return 6

def replace_distribution(d):
    """
    Esta función remplaza los valores por numericos
    """
    if d == "X":
        return 1
    elif d == "O":
        return 2
    elif d == "I":
        return 3
    elif d == "C":
        return 4

def replace_classFlare(cf):
    if cf == "M":
        return 0
    elif cf == "X":
        return 1
```

```python
lista = ['class','penumbra','distribution']
for i in lista:
    exec('df["'+i+'"] = df["'+i+'"].apply(lambda x: replace_'+i+'(x))')
```

```python
df['class_flare']=df['class_flare'].apply(lambda x: replace_classFlare(x))
```

# Bayesian Logistic Regression

$$ y_i | \pi_i \sim Bernoulli(p) $$

$$ \begin{split}Y =
\left\{
    \begin{array}{ll}
        1 & \textrm{Si es una erupción clase X} \\
        0 & \textrm{Si es una erupción clase M}
    \end{array}
\right.\end{split} $$

$$ \pi_i = P\{y_i = 1 \} = \frac{exp(x^t_i * \beta)}{1+ exp(x^t_i * \beta)} $$

## Definir priors & family

```python
# Likelihood
likelihood = bmb.Likelihood("Bernoulli", parent="p")

# Regresión Logistica, para las Y_i
family = bmb.Family("bernoulli", likelihood, "logit")
```

```python
pr = bmb.Prior('Normal', mu=0.5, sigma=500)
dict_prior = {'class':pr, 'penumbra':pr,'distribution':pr,'activity':pr,'evolution':pr,'flare24.activity':pr,
             'history':pr,'sun.disk':pr}
```

```python
model=bmb.Model('class_flare ~ class+penumbra+distribution+activity+evolution+flare24.activity+history+sun.disk',df,
                family=family,priors=dict_prior)
model
```

```python
# Fit the model using 1000 on each of 3 chains
results = model.fit(draws=1000, chains=3, method='mcmc')
```
